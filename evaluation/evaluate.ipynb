```python
import os
import json
import pandas as pd
from datasets import Dataset
from ragas import evaluate
from ragas.metrics import (
    answer_relevancy,
    context_precision,
    context_recall,
    faithfulness,
    answer_similarity
)
from openai import OpenAI

# 1. إعداد مفتاح OpenAI (تأكد أنك خزّنت المفتاح في بيئة التشغيل)
os.environ["OPENAI_API_KEY"] = "----"  # ← ضع التوكن الخاص بك هنا

# 2. تحميل الملف بصيغة JSON
with open("/content/rag_evaluate_question_with_reference.json", "r", encoding="utf-8") as f:
    data = json.load(f)

# 3. تحويل البيانات إلى صيغة DataFrame
df = pd.DataFrame(data)

# 4. التأكد أن الأعمدة تتضمن السياقات على شكل قائمة
df["contexts"] = df["contexts"].apply(lambda x: x if isinstance(x, list) else [x])

# 5. تحويل DataFrame إلى HuggingFace Dataset
dataset = Dataset.from_pandas(df)

# 6. تقييم باستخدام RAGAS
result = evaluate(
    dataset=dataset,
    metrics=[
        answer_relevancy,
        context_precision,
        context_recall,
        faithfulness,
        answer_similarity
    ]
)

# 7. طباعة النتائج
print(result)
```
